#a. List vs. Tuple
Mutability:

List: Mutable (can be changed after creation).
Tuple: Immutable (cannot be changed after creation).
Syntax:

List: Defined using square brackets [].
Tuple: Defined using parentheses ().
Performance:

List: Generally, takes up slightly more memory and is a bit slower than a tuple due to its flexibility.
Tuple: Takes up less memory and is faster for indexed lookup and iteration compared to a list, especially for large data.
Use-Cases:

List: Useful when you need a sequence type that may need to be modified.
Tuple: Useful when you need a read-only (hashable) data structure.
Hashability:

List: Cannot be used as a key in a dictionary because it is unhashable.
Tuple: Can be used as a key in a dictionary (if it contains only hashable elements).
b. Dictionary vs. Set
Element Type:

Dictionary: Consists of key-value pairs.
Set: Consists of unique elements.
Syntax:

Dictionary: Defined with curly braces and uses colons to separate keys and values, e.g., {key1: value1, key2: value2}.
Set: Defined with curly braces but contains single elements, e.g., {element1, element2}.
Order:

Dictionary: As of Python 3.7, dictionaries maintain the insertion order of items.
Set: Does not maintain the order of elements.
Usage:

Dictionary: Ideal when you need a logical association between a key:value pair.
Set: Useful when you need a collection with no duplicate elements.
Lookups:

Dictionary: Allows looking up, modifying, and retrieving elements using a key.
Set: Allows checking whether an element is part of the set, but does not allow direct retrieval of an element.
Mutable Variants:

Dictionary: No immutable variant in the Python standard library.
Set: Has an immutable variant called a "frozenset".
These are some of the fundamental differences between the data structures. Depending on your specific use case and operations needed, you would choose the most fitting data structure.


Generate random numbers (integer) 
Generate 5 random integers and calculate the sum 
Find the max of the number
Find the min of the number

```python
import random

# Generate 5 random integers between a desired range, e.g., 1 and 100
random_numbers = [random.randint(1, 100) for _ in range(5)]

# Calculate the sum of the numbers
sum_numbers = sum(random_numbers)

# Find the maximum number
max_number = max(random_numbers)

# Find the minimum number
min_number = min(random_numbers)

# Output the results
print("Random numbers: ", random_numbers)
print("Sum: ", sum_numbers)
print("Max: ", max_number)
print("Min: ", min_number)
```

```python
Give an example for each of the following. 
Addtionally use three built in functions in python - given for each of them 

a. list
b. Tuple
c. Set
d. Dictionary
```

```python
a. List
# Example List
my_list = [1, 3, 5, 3, 7, 3]

# Built-in function 1: len() - Returns the length of the list
print(len(my_list))  # Output: 6

# Built-in function 2: sum() - Returns the sum of all elements in the list
print(sum(my_list))  # Output: 22

# Built-in function 3: sorted() - Returns a new list containing all elements from the original list in ascending order
print(sorted(my_list))  # Output: [1, 3, 3, 3, 5, 7]

b. Tuple
# Example Tuple
my_tuple = (10, 20, 30, 40)

# Built-in function 1: len() - Returns the number of items in the tuple
print(len(my_tuple))  # Output: 4

# Built-in function 2: min() - Returns the item with the lowest value in the tuple
print(min(my_tuple))  # Output: 10

# Built-in function 3: max() - Returns the item with the highest value in the tuple
print(max(my_tuple))  # Output: 40

c. Set
# Example Set
my_set = {3, 7, 11, 15}

# Built-in function 1: len() - Returns the number of elements in the set
print(len(my_set))  # Output: 4

# Built-in function 2: min() - Returns the smallest item in the set
print(min(my_set))  # Output: 3

# Built-in function 3: max() - Returns the largest item in the set
print(max(my_set))  # Output: 15

d. Dictionary
# Example Dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3}

# Built-in function 1: len() - Returns the number of items (key-value pairs) in the dictionary
print(len(my_dict))  # Output: 3

# Built-in function 2: keys() - Returns a view object that displays a list of all the keys in the dictionary
print(my_dict.keys())  # Output: dict_keys(['a', 'b', 'c'])

# Built-in function 3: values() - Returns a view object that displays a list of all the values in the dictionary
print(my_dict.values())  # Output: dict_values([1, 2, 3])


```

```python
List the differences between in and is operator and use it in 
a. list
b. Tuple
c. Set
d. Dictionary
```

```python
#In Python, in and is are two different operators used for various kinds of operations:

#in Operator:
#Purpose: Checks if a value is present in a given sequence (like lists, strings, tuples) or mapping (like dictionaries).
#Type: Membership operator.
#is Operator:
#Purpose: Checks if two variables point to the same object (i.e., have the same memory address).
#Type: Identity operator.

# Example Usage:
# a. List
#Using in:

my_list = [1, 2, 3, 4]
print(3 in my_list)  # Output: True

# Using is:
list1 = [1, 2, 3]
list2 = list1
print(list1 is list2)  # Output: True

# b. Tuple
# Using in:
my_tuple = (1, 2, 3, 4)
print(5 in my_tuple)  # Output: False

# Using is:
tuple1 = (1, 2, 3)
tuple2 = tuple1
print(tuple1 is tuple2)  # Output: True

# c. Set
# Using in:
my_set = {1, 2, 3, 4}
print(3 in my_set)  # Output: True

# Using is:

set1 = {1, 2, 3}
set2 = set1
print(set1 is set2)  # Output: True

# d. Dictionary
# Using in:
my_dict = {'a': 1, 'b': 2, 'c': 3}
print('b' in my_dict)  # Output: True (checks keys, not values)

# Using is:
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = dict1
print(dict1 is dict2)  # Output: True

# The in operator is often used for iteration, containment checking, and also to check if a key exists in a dictionary.
# The is operator is used for identity checks, and it's not recommended to use it for equality checks for primitive types 
# (like checking if two strings or numbers are the same), due to the way Python manages its memory (interning and caching mechanisms). Use == for value equality checks instead.




```
